## Pre-Processing the Data

### Year

```python
# Convert to absolute values
imdb_df['Year'] = imdb_df['Year'].abs()

# Convert from float to integer
imdb_df['Year'] = imdb_df['Year'].astype(int)
```

### Duration

```python
# Remove the 'min' suffix and convert to float
imdb_df['Duration'] = imdb_df['Duration'].str.replace(' min', '').astype(float)

# Convert the column to integer
imdb_df['Duration'] = imdb_df['Duration'].astype(int)
```

### Genre

```python
# Splittig the genre by "," and filling the NaN with the mode
imdb_df['Genre'] = imdb_df['Genre'].str.split(', ')
imdb_df = imdb_df.explode('Genre')
imdb_df['Genre'].fillna(imdb_df['Genre'].mode()[0], inplace = True)
```

### Votes

```python
# Remove commas from the 'Votes' column and convert it to integer
imdb_df['Votes'] = imdb_df['Votes'].str.replace(',', '').astype(int)
```

## Data visualizing

### Year

```python
year = px.histogram(imdb_df, x = 'Year', histnorm = 'probability density', nbins = 30)
year.show()
```

```python
# Group data by year and calculate the average rating
avg_rating_by_year = imdb_df.groupby(['Year', 'Genre'])['Rating'].mean().reset_index()
```

```python
# Get the top 10 genres
top_genres = imdb_df['Genre'].value_counts().head(10).index
```

```python
# Filter the data to include only the top 10 genres
avg_rating_by_year = avg_rating_by_year[avg_rating_by_year['Genre'].isin(top_genres)]
```

```python
# Create the line plot
fig = px.line(avg_rating_by_year, x = 'Year', y = 'Rating', color = "Genre")
fig
```

```python
fig.update_layout(title = 'Average Rating by year for Top Genres', xaxis_title = 'Year', yaxis_title = 'Average Rating')
fig.show()
```

```python
# Probability density of Rating

rating_fig = px.histogram(imdb_df, x = 'Rating', histnorm = 'probability density', nbins = 40)
rating_fig.update_layout(title = 'Distribution of Rating', title_x = 0.5, title_pad = dict(t=20), title_font = dict(size=20), xaxis_title = 'Rating', yaxis_title = 'Probability density')
rating_fig.show()
```
